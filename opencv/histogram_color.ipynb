```python
import cv2
import numpy as np
import imutils
from matplotlib import pyplot as plt

```

```python
image = cv2.imread('../resources/car.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# split image into 3 channels
channels = cv2.split(image)
colors = ['b', 'g', 'r']

plt.figure()
plt.title('Flattened color histogram')
plt.xlabel('Bins')
plt.ylabel('number of pixels')

for channel, color in zip(channels, colors):
    hist = cv2.calcHist([channel], [0], None, [256], [0,256])
    plt.plot(hist, color)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_title('2d histogram of green and blue')
hist = cv2.calcHist([channels[1], channels[0]], [0,1], None, [32, 32], [0,256, 0,256])
p = ax.imshow(hist)
plt.colorbar(p)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_title('2d histogram of green and red')
hist = cv2.calcHist([channels[1], channels[2]], [0,1], None, [32, 32], [0,256, 0,256])
p = ax.imshow(hist)
plt.colorbar(p)
print(hist.flatten())
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_title('2d histogram of blue and red')
hist = cv2.calcHist([channels[0], channels[2]], [0,1], None, [32, 32], [0,256, 0,256])
p = ax.imshow(hist)
plt.colorbar(p)
```

```python
print(hist.shape)
print(hist.flatten()[0])
```

```python
# 3d histogram
hist = cv2.calcHist([image], [0,1,2], None, [8,8,8], [0,256,0,256,0,256])
print(hist.shape)
print(hist.flatten()[:5])

plt.figure()
plt.axis('off')
plt.imshow(imutils.opencv2matplotlib(image))
plt.show()
```
