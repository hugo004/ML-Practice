```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
```

```python
# color boundaries definition
boundaries = [
    ([136, 87, 111], [180, 255, 255]), # red color
    ([25, 52, 72], [102, 255, 255]), # green color
    ([94, 80, 2], [120, 255, 255]) # blue color,
]


image = cv2.imread('../resources/color_transfer_sample2.jpg')
orig = image.copy()
# image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
outputs = []
for lower, upper in boundaries:
    lower = np.array(lower, dtype=np.uint8)
    upper = np.array(upper, dtype=np.uint8)
    
    mask = cv2.inRange(image, lower, upper)
    output = cv2.bitwise_and(image, image, mask=mask)
    
    outputs.append(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))

plt.figure(figsize=(12,10))
plt.imshow(np.vstack(outputs))
    
    
```
