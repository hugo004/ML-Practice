```python
import cv2
import matplotlib.pyplot as plt
import skimage
```

```python
img = cv2.imread("../../datasets/cat.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
resized_img = cv2.resize(img, (825, 925))
plt.imshow(resized_img)
```

```python
_, hog_img = skimage.feature.hog(resized_img, cells_per_block=(5, 5), visualize=True, multichannel=True)
plt.imshow(hog_img)
```

We can use "skimage" to extract HOG features easily. We can use the feature to solve classification problems.
