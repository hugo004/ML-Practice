```python
import pandas as pd
import pickle
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.pipeline import make_pipeline

```

```python
# df = pd.read_csv('../../hand_gesture_coords.csv')
df = pd.read_csv('../../hand_gesture_coords_counting.csv')
df.head()
```

```python
le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])
le.classes_
```

```python
df.head()
```

```python
X = df.drop('class', axis=1)
y = df['class']

X_train, X_valid, y_trian, y_valid = train_test_split(X, y, test_size=0.3, random_state=44)
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC


# model = make_pipeline(StandardScaler(), LogisticRegression())
# model = make_pipeline(StandardScaler(), KNeighborsClassifier(5))
# model = make_pipeline(StandardScaler(), DecisionTreeClassifier(max_depth=10))
# model = make_pipeline(StandardScaler(), GaussianNB())
# model = make_pipeline(StandardScaler(), RandomForestClassifier(max_depth=10))
# model = make_pipeline(StandardScaler(), GaussianProcessClassifier(1.0 * RBF(1.0)))
# model = make_pipeline(StandardScaler(), AdaBoostClassifier())
# model = make_pipeline(StandardScaler(), MLPClassifier())
# model = make_pipeline(StandardScaler(), SVC(gamma=2, C=1))
# model = make_pipeline(StandardScaler(), SGDClassifier(loss='log', shuffle=True, random_state=44))
model = make_pipeline(StandardScaler(), GradientBoostingClassifier(random_state=44, learning_rate=0.01))


model.fit(X_train, y_trian)

pred = model.predict(X_valid)
accuracy_score(y_valid, pred, normalize=True)
```

```python
joblib.dump(model, 'hand_gesture_counting.joblib')
```
