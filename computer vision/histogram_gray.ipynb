```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('../resources/card2.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# 0 channel means gray scale image
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
plt.figure()
plt.axis('off')
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_GRAY2BGRA))
plt.show()
```

```python
plt.figure()
plt.title('Gray scale histogram')
plt.xlabel('Bins')
plt.ylabel('count of pixels')
plt.plot(hist)
# plt.xlim([0, 256])
```

```python
# reference: https://www.youtube.com/watch?v=Vq-egjXo4Yc&list=PLACfP48L5l2PZLhn6QLNzjVR4EQ3z_EPo&index=7
# normalize, after the normalize the value of sum will be 1
# it can avoid the size issue between different images
# e.g large image contains more pixels, small image contains less pixels
hist /= hist.sum()
```

```python
plt.figure()
plt.title('gray scale histogram (normalized')
plt.xlabel('bins')
plt.ylabel('% of pixels')
plt.plot(hist)
plt.show()
```
