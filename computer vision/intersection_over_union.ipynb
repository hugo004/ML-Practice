```python
import cv2
import numpy as np

from matplotlib import pyplot as plt
```

```python
image = cv2.imread('../datasets/object_detect_sample.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(image_rgb)
```

```python
copy = image_rgb.copy()
cv2.rectangle(copy, (130, 40), (430, 400), (0, 255, 0), 3)
cv2.rectangle(copy, (200, 100), (450, 420), (255, 0, 0), 3)

plt.imshow(copy)
```

```python
def bounding_box_iou(a, b):
    # compute the intersection area of rectangle
    # the inner coordinates
    ax = max(a[0], b[0])
    ay = max(a[1], b[1])
    bx = min(a[2], b[2])
    by = min(a[3], b[3])
    
    # w * h = area
    inter_area = max(0, bx - ax + 1) * max(0, by - ay + 1)
    area_a = (a[2] - a[0] + 1) * (a[3] - a[1] + 1)
    area_b = (b[2] - b[0] + 1) * (b[3] - b[1] + 1)
    
    # iou formula
    iou = inter_area / float(area_a + area_b - inter_area)
    
    return iou
```

```python
ground_true = [130, 40, 430, 400]
predict = [200, 100, 450, 420]

iou = bounding_box_iou(ground_true, predict)
iou
```
