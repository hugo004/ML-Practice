```python
import cv2
import numpy as np
import imutils
from matplotlib import pyplot as plt

```

```python
image = cv2.imread('../resources/car.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# reference: https://www.youtube.com/watch?v=r-xPx_axcok&list=PLACfP48L5l2PZLhn6QLNzjVR4EQ3z_EPo&index=9
plt.figure()
mask = np.zeros(image.shape[:2], dtype='uint8')
cv2.rectangle(mask, (600, 550), (210, 390), 255, -1)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.figure()
masked = cv2.bitwise_and(image, image, mask=mask)
plt.imshow(cv2.cvtColor(masked, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# origin histogram
channels = cv2.split(image)
colors = ['b', 'g', 'r']

plt.figure()
plt.title('Flattened color histogram')
plt.xlabel('Bins')
plt.ylabel('number of pixels')

for channel, color in zip(channels, colors):
    hist = cv2.calcHist([channel], [0], None, [256], [0,256])
    plt.plot(hist, color)
```

```python
# split image into 3 channels
channels = cv2.split(image)
colors = ['b', 'g', 'r']

plt.figure()
plt.title('Flattened color histogram with mask')
plt.xlabel('Bins')
plt.ylabel('number of pixels')

# only check histogram of the mask 
for channel, color in zip(channels, colors):
    hist = cv2.calcHist([channel], [0], mask, [256], [0,256])
    plt.plot(hist, color)
```
